# How to write equations in matplotlib.pyplot

Practice makes perfect

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = r'3 = \frac{12}{4}'
ax = plt.axes([0,0,0.3,0.3]) #left,bottom,width,height
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
plt.text(0.4,0.4,'$%s$' %a,size=50,color="green")
print('Source: ','https://matplotlib.org/3.1.1/gallery/text_labels_and_annotations/mathtext_examples.html')
```

```python
a = r'16 = {8}*{2}'
ax = plt.axes([0,0,0.3,0.3]) #left,bottom,width,height
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
plt.text(0.4,0.4,'$%s$' %a,size=50,color="green")
print('source: ','https://matplotlib.org/3.1.1/gallery/text_labels_and_annotations/mathtext_examples.html')
print('We create a mathematical formula in Python:')
```
