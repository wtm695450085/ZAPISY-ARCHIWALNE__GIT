```python
import seaborn as sns

sns.pairplot(data=df[['age', 'duration', 'campaign', 'y']], hue='y', dropna=True)
```

```python
sns.pairplot(data=df[['age', 'duration', 'campaign', 'y']], hue='y', dropna=True, height=2.5, palette=('husl'), markers='x')
```

```python
sns.pairplot(data=df[['age', 'duration', 'campaign', 'y']], hue='y', dropna=True, height=2.5, palette=('BrBG'), markers='x')
```
